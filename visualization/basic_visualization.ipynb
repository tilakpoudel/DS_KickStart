# Univariate Visulalization

Univarite alalysis looks at one feature at a time. When we analyze feature independently, we are usually mostly interested in the feature independently, we are usually interested in the distribution of its values and ignore other features in the dataset.

## Histogram and density plot


It helps to know the distribution of the data .

```python
features = ['RHI_Infant_Mortality_Ind','RHI_Premature_Ind']
df[features].describe()
```

```python
print(df[features].head())
df[features].hist(figsize=(10,4))
```

```python
help(df.plot())
```

```python
df[features].plot(kind='density', subplots=True, layout=(1,2),sharex=False, figsize=(10,4));
```

Plotting using snsplot

```python
sns.distplot(df['RHI_Infant_Mortality_Ind']);
```

## Box plot

```python
sns.boxplot(x='RHI_Infant_Mortality_Ind',data=df);
```

## Violine plot

The difference between the box and violin plots is that the former illustrates certain statistics concerning individual examples in a dataset while the violin plot concentrates more on the smoothed distribution as a whole.

```python
_, axes = plt.subplots(1,2, sharey=True,figsize=(6,4));
sns.boxplot(data=df['RHI_Infant_Mortality_Ind'],ax=axes[0]);
sns.violinplot(data=df['RHI_Infant_Mortality_Ind'],ax=axes[1]);
```
