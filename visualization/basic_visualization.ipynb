# Univariate Visulalization

Univarite alalysis looks at one feature at a time. When we analyze feature independently, we are usually mostly interested in the feature independently, we are usually interested in the distribution of its values and ignore other features in the dataset.

## Histogram and density plot


It helps to know the distribution of the data .

```python
features = ['RHI_Infant_Mortality_Ind','RHI_Premature_Ind']
df[features].describe()
```

```python
print(df[features].head())
df[features].hist(figsize=(10,4))
```

```python
help(df.plot())
```

```python
df[features].plot(kind='density', subplots=False, layout=(1,2),sharex=True, figsize=(10,4));
```

Plotting using snsplot

```python
sns.distplot(df['RHI_Infant_Mortality_Ind']);
```

## Box plot

```python
sns.boxplot(x='RHI_Infant_Mortality_Ind',data=df);
```

## Violine plot

The difference between the box and violin plots is that the former illustrates certain statistics concerning individual examples in a dataset while the violin plot concentrates more on the smoothed distribution as a whole.

```python
_, axes = plt.subplots(1,2, sharey=True,figsize=(6,4));
sns.boxplot(data=df['RHI_Infant_Mortality_Ind'],ax=axes[0]);
sns.violinplot(data=df['RHI_Infant_Mortality_Ind'],ax=axes[1]);
```

```python
np.random.seed(1)
x = np.arange(5)
y = np.random.randn(5)

fig, axes = plt.subplots(ncols=2, figsize=plt.figaspect(1))

vert_bars = axes[0].bar(x, y, color='lightblue', align='center')
horiz_bars = axes[1].barh(x, y, color='lightblue', align='center')

# I'll also introduce axhline & axvline to draw a line all the way across the axes
# This can be a quick-n-easy way to draw an axis "spine".
axes[0].axhline(0, color='gray', linewidth=2)
axes[1].axvline(0, color='gray', linewidth=2)

plt.show()
```

## Filled Region

```python
np.random.seed(1)
y = np.random.randn(100).cumsum()
x = np.linspace(0, 10, 100)
print(y,x)
fig, ax = plt.subplots()
ax.fill_between(x, y, color='lightblue')
plt.show()
```

## Scatter plot

```python
x=df['RHI_Infant_Mortality_Ind']
y= df['RHI_Premature_Ind']
plt.scatter(x,y,color='green')
plt.title('RHI_Premature_Ind vs RHI_Infant_Mortality_Ind')
plt.xlabel('RHI_Infant_Mortality_Ind')
plt.ylabel('RHI_Premature_Ind')
plt.show()
```
