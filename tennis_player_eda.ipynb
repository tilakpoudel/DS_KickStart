```python
world['Total_Professional_Matches']= world.Career_wins + world.Career_losses
# world.head()
world['Win_Percentage']=(world.Career_wins / world.Total_Professional_Matches)* 100
world[['Tennis_Player','Total_Professional_Matches','Win_Percentage']].sort_values('Win_Percentage',ascending=False).head(10)
```

Getting the percentage difference between each player after sorting

```python
tennis_win_perc = world.sort_values('Win_Percentage',ascending=False)
tennis_win_perc['Percent_diff'] = tennis_win_perc.Win_Percentage.pct_change()*100
tennis_win_perc['Percent_diff']= tennis_win_perc.Percent_diff.fillna(0)
tennis_win_perc[['Tennis_Player', 'Win_Percentage', 'Percent_diff']].head(10)
```

Filtering with Easy Readability


```python
crit1 = world.Maximum_Consectutive_Weeks_at_no_1 > 30
crit2 = world.Titles > 80
crit3 = world.Prize_Money < 100000000
```

```python
world[crit1 & crit2 & crit3].loc[:,['Tennis_Player']]
```

The reduce function from the functools module works similarly with a lambda expression, by taking the filter product from the first 2 criteria and combining them with the third to produce a single filtering condition that can be applied to the DataFrame.

```python
from functools import reduce
criteria = reduce(lambda x,y:x & y, [crit1, crit2, crit3])
```

```python
world[criteria].loc[:,['Tennis_Player']]
```

# Informative Scatter Plots
Exploratory scatter plots can be used to investigate the relationship between two continuous variables. Here I simply create lists of the tennis players, their Earnings, and their win percentage. I then create a for loop that iterates through these three lists and plots them, creating an informative scatter plot.
N.B. The Earnings column was created for this scatter plot, as shown via the code snippet.

```python
world['Earnings']= tennis_win_perc.Prize_Money/ 10 ** 6
```

```python
tennis_players = list(world.Tennis_Player)
x_coords=list(world.Earnings)
y_coords = list(world.Win_Percentage)

plt.figure(figsize =(18,12))

for i, player in enumerate(tennis_players):
    x= x_coords[i]
    y= y_coords[i]
    plt.scatter(x,y, marker='o',color='red',alpha=0.5,s=100)
    plt.text(x+0.3 ,y+0.3,player,fontsize=10)
    
plt.title('The Relationship between Win Percentage and Prize Money\nfor Former World Number No.1s', fontsize=20)
plt.ylabel('Win Percentage (%)', fontsize=16)
plt.xlabel('Prize Money', fontsize=16)
plt.show()
```
